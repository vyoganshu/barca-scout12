```python
import pandas as pd
```

```python
df = pd.read_csv("players_data_light-2024_2025.csv")
df
```

```python
df = df[["Player","Nation","Pos","Age","PrgC","PrgP","PrgR","G+A","xG+xAG","Min","Squad"]]
df
```

```python
squads = ["Inter","Chelsea","Arsenal","Juventus","Manchester City","Liverpool","Paris S-G",
          "Tottenham","Bayern Munich","Milan",
          "Leverkusen","Juventus",
          "Real Madrid","Manchester Utd","Crystal Palace","Napoli","Valencia",
          "Real Sociedad","Roma","Marseille","Lazio",
          "Dortmund","Mallorca","Sevilla"]
positions = ["FW","MF"]

df = df[(df['Min'] > 1100) & (df['Squad'].isin(squads)) & (df['Pos'].isin(positions))]
df
```
