```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi
import pandas as pd
```

PEDRI v FRATTESI

```python
df = pd.read_csv("new_all_players.csv")
players = ["Harry Kane", "Raphinha"]
df = df[df["Player"].isin(["Harry Kane", "Raphinha"])].copy()

df["PrgC_per90"] = df["PrgC"] / (df["Min"] / 90)
df["PrgP_per90"] = df["PrgP"] / (df["Min"] / 90)
df["PrgR_per90"] = df["PrgR"] / (df["Min"] / 90)
df["GA_per90"]   = df["G+A"] / (df["Min"] / 90)
df["xGxAG_per90"] = df["xG+xAG"] / (df["Min"] / 90)

stats_to_plot = ["PrgC_per90", "PrgP_per90", "PrgR_per90", "GA_per90", "xGxAG_per90"]
plot_data = df[["Player"] + stats_to_plot].melt(id_vars="Player", var_name="Stat", value_name="Per90")

plt.figure(figsize=(6, 6))
sns.barplot(x="Stat", y="Per90", hue="Player", data=plot_data)
plt.title("Harry Kane vs Raphinha: Per 90 Stats")
plt.xlabel("Stat")
plt.ylabel("Per 90 Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ---- SCATTER PLOT ----
df["PrgP_per90"] = df["PrgP"] / (df["Min"] / 90)
df["GA_per90"] = df["G+A"] / (df["Min"] / 90)

plt.figure(figsize=(8, 6))
plt.scatter(df["PrgP_per90"], df["GA_per90"], alpha=0.3, label="Other Players")

for _, row in players.iterrows():
    plt.scatter(row["PrgP_per90"], row["GA_per90"], label=row["Player"], s=100, marker="X")
    plt.text(row["PrgP_per90"] + 0.05, row["GA_per90"] + 0.05, row["Player"])

plt.xlabel("Progressive Passes per 90")
plt.ylabel("G+A per 90")
plt.title("Scatter Plot: PrgP vs G+A per 90")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("new_all_players.csv")

# Clean up
df = df[df["Min"] > 0].copy()

# Create per 90 stats
df["PrgC_per90"]   = df["PrgC"] / (df["Min"] / 90)
df["PrgP_per90"]   = df["PrgP"] / (df["Min"] / 90)
df["PrgR_per90"]   = df["PrgR"] / (df["Min"] / 90)
df["GA_per90"]     = df["G+A"] / (df["Min"] / 90)
df["xGxAG_per90"]  = df["xG+xAG"] / (df["Min"] / 90)

# Setup your comparison variables
x_stat = "PrgP_per90"
y_stat = "GA_per90"

# Choose players
highlight = {
    "Jude Bellingham": "green",
    "Frenkie de Jong": "red"
}

# Base scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=x_stat, y=y_stat, color="lightblue", alpha=0.6, label="Other Players")

# Highlight players
for player, color in highlight.items():
    temp = df[df["Player"] == player]
    if not temp.empty:
        plt.scatter(temp[x_stat], temp[y_stat], color=color, s=120, label=player, marker='X', zorder=3)
        plt.text(temp[x_stat].values[0] + 0.2, temp[y_stat].values[0], player, fontsize=9, weight='bold')

# Labels and formatting
plt.title(f"Scatter Plot: {x_stat} vs {y_stat}", fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
